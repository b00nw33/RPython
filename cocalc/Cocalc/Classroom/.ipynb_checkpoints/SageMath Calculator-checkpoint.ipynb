### Pascal's Triangle

```python
[[binomial(n,r) for r in range(n+1)] for n in range(8)]
```

The diagonals of the Pascal's Triangle contain coefficients of linear, quadratic, cubic polynomials etc, that can be useful for the study of polynomial sequences. Take diagonal column 3 (note that n starts from 0) for example:

```python
n = 3
col3 = [[binomial(n+i,n) for i in range(8)]]
col3
```

```python
import numpy as np
col3.append(list(np.diff(col3[-1])))
col3
```

Since there exists 3 non-zero difference rows, a 3rd deg. polynomial can be used to generate the diagonal column 3.

```python
a,b,c,d,x = var('a,b,c,d,x')

f(x) = a*x^3 + b*x^2 + c*x + d
d0 = [[f(n) for n in range(1,5)]]
d0
```

```python
d0.append(list(np.diff(d0[-1])))
[i[0] for i in d0]
```

```python
[f.subs(a=1/6,b=1/2,c=1/3,d=0)(n) for n in range(1,6)]
```

### Kinematics equations (SUVAT)

```python
s,u,v,a,t = var('s,u,v,a,t')

# SUVAT Equations
uvat = v == u + a*t
uvas = v**2 == u**2 + 2*a*s
utas = s == u*t + .5*a*t**2
vuts = s == (v+u)/2*t
svat = s == v*t - .5*a*t**2

sol = solve(uvas.subs(a=-9.81, v=0, s=91.5), u)

R = RealField(24)
[R(i.rhs()) for i in sol]
```

### Grading of results

```python
def gradeN(score):
    grades = 'A','B','C','D','E','U'
    bounds = 75,70,65,60,50,0
    # return(grades[min((k for k,v in enumerate(bounds) if score>=v))])
    # return(grades[len(list(filter(lambda x: x > score, bounds)))])
    return(grades[sum([score<bound for bound in bounds])])

list(map(gradeN,[75,70,65,60,50,0]))
```

### Generators

```python
# https://www.learnpython.org/en/Generators

def fib():
    a,b = 1,1
    while 1:
        yield a
        a,b = b,a+b

# [i for i in fib(10^6)]

count=0
for i in fib():
    print(i)
    count+=1
    if count==5: break
```

```python
def fact(m):
    def gen(n):
        i = n
        while i>0:
            yield i
            i-=1
    return prod(i for i in gen(m))

fact(50)
```

```python
import math

print('{:^3} {:^6} {:6} {:6}'.format(
    'e', 'x', 'x**2', 'isinf'))
print('{:-^3} {:-^6} {:-^6} {:-^6}'.format(
    '', '', '', ''))

for e in range(0, 201, 20):
    x = 10.0 ** e
    y = x * x
    print('{:3d} {:<6g} {:<6g} {!s:6}'.format(
        e, x, y, math.isinf(y),
    ))
```

```python
import math

INPUTS = [
    (1000, 900, 0.1),
    (100, 90, 0.1),
    (10, 9, 0.1),
    (1, 0.9, 0.1),
    (0.1, 0.09, 0.1),
]

print('{:^8} {:^8} {:^8} {:^8} {:^8} {:^8}'.format(
    'a', 'b', 'rel_tol', 'abs(a-b)', 'tolerance', 'close')
)
print('{:-^8} {:-^8} {:-^8} {:-^8} {:-^8} {:-^8}'.format(
    '-', '-', '-', '-', '-', '-'),
)

fmt = '{:8.2f} {:8.2f} {:8.2f} {:8.2f} {:8.2f} {!s:>8}'

for a, b, rel_tol in INPUTS:
    close = math.isclose(a, b, rel_tol=rel_tol)
    tolerance = rel_tol * max(abs(a), abs(b))
    abs_diff = abs(a - b)
    print(fmt.format(a, b, rel_tol, abs_diff, tolerance, close))
```

```python
# math_trig.py

import math

print('{:^7} {:^7} {:^7} {:^7} {:^7}'.format(
    'Degrees', 'Radians', 'Sine', 'Cosine', 'Tangent'))
print('{:-^7} {:-^7} {:-^7} {:-^7} {:-^7}'.format(
    '-', '-', '-', '-', '-'))

fmt = '{:7.2f} {:7.2f} {:7.2f} {:7.2f} {:7.2f}'

for deg in range(0, 361, 30):
    rad = math.radians(deg)
    if deg in (90, 270):
        t = float('inf')
    else:
        t = math.tan(rad)
    print(fmt.format(deg, rad, math.sin(rad), math.cos(rad), t))
```

```python
# math_log10.py
import math

print('{:2} {:^15} {:^20} {:^25} {:8}'.format(
    'i', 'x', 'accurate', 'inaccurate', 'mismatch',
))
print('{:-^2} {:-^15} {:-^20} {:-^25} {:-^8}'.format(
    '', '', '', '', '',
))

for i in range(0, 10):
    x = math.pow(10, i)
    accurate = math.log10(x)
    inaccurate = math.log(x, 10)
    match = '' if int(inaccurate) == i else '*'
    print('{:2d} {:15.1f} {:^20.15f} {:^25.18f} {:^8}'.format(
        i, x, accurate, inaccurate, match,
    ))
```
