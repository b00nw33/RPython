## Lecture 4: Limits and Derivatives



```python
f(x) = x/abs(x)

plot(f, figsize=4)
```

```python
print(limit(f, x=0))
print(limit(f, x=0, dir='+'))
print(limit(f, x=0, dir='-'))
```

```python
g(x) = (x**2-4)/(x-2)

plot(g, 0, 5, figsize=4)
```

```python
print(limit(g, x=2))
print(limit(g, x=2, dir='+'))
print(limit(g, x=2, dir='-'))

# Limit exists at x=2, even though x is undefined. Therefore, g(x) is continuous at x=2.
```

```python
# Limits of sequences

var('n')
f(n) = (1 + 1/n)**n

limit(f, n=oo)
```

```python
f(n) = 5**(1/n) - 1

N = 50

p = points([(n,f(n)) for n in range(1,N)], color='red')
p1 = points([(n,1/n) for n in range(1,N)], alpha=.3)
p2 = points([(n,-1/n) for n in range(1,N)], alpha=.3)

show(p+p1+p2, figsize=4)

# Find the limits!
```

```python
# Recursive sequence

var('n')

def Y(n):
    if n==1: return 1
    else: return (Y(n-1) + 2/Y(n-1)) / 2

Y(15).n()
```

```python
var('n')

def memoized_Y(n, D={}):
    """
    Returns the n-th term using D to memoize
    the values computed in previous calls.
    """

    if n in D.keys():
        return D[n]
    else:
        if n==1: result=1
        else: result= (memoized_Y(n-1) + 2/memoized_Y(n-1)) /2
    D[n] = result
#     print(D)
    return result

memoized_Y(15).n()
```

```python
# Finding derivatives using first principles

a, x = var('a,x')
h(x) = log(x) + x**5 + sin(x)

limit((h(x+a) - h(x)) / a, a=0)
```

```python
h.diff()
```
