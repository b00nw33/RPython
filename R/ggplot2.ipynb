# `qplot()` vs `ggplot()`
Plots for teaching of Maths and Science. Let's start by exploring `qplot()`, which is basically the shorthand syntax of `ggplot()`.

```r
# Count of CAT
library(ggplot2)

data = ceiling(runif(200,0,3))

tab = table(data)
tab
prop.table(table(data))*100
avg = sum(as.integer(names(tab)) * tab) /sum(tab)
cat('Average =', avg)
```

```r
qplot(data, geom=c('bar', 'histogram'), alpha=I(.2))
```

```r
# CAT vs CAT
library(ggplot2)

df = data.frame(A=ceiling(runif(400,0,3)),B=ceiling(runif(400,3,6)))

table(df)
prop.table(table(df))*100
```

```r
# Count of QUANT
library(ggplot2)

sex = rep(c('F','M'), c(100,500))  # Replace c(100,500) with each=500
height = c(rnorm(100,140,20), rnorm(500,180,20))
df = data.frame(sex, height)
df$sex = factor(df$sex, levels=c('M','F'), ordered=T)

# options(digits=4)
# df

str(df)
summary(df)
round(quantile(df$height,c(.25,.5,.75)),2)

# hist() and density() in base R
# Note: Histogram bins are STACKED, density plots illustrate proportion of individual CAT
qplot(height, data=df, geom='histogram', fill=sex, alpha=I(.2))
qplot(height, data=df, geom='density', fill=sex, alpha=I(.2))
```

```r
# Ogive with quartile levels

plot(ecdf(df$height)) +
abline(h=c(.25,.5,.75), v=c(fivenum(df$height)[2:4]), lty=2)
```

```r
# CAT vs QUANT
qplot(sex, height, data=df, geom=c('boxplot','jitter'), fill=sex, color=sex, alpha=I(.2))
```

```r
# QUANT vs QUANT
library(ggplot2)

x = -5:5
y = (x-1)*(x+2)
df = data.frame(x,y)

qplot(x,y, data=df, geom=c('point','line')) +
geom_hline(yintercept = 0, lty=2) +
geom_vline(xintercept = 0, lty=2)
```

# `ggplot()`

```r
# Count of CAT
library(ggplot2)

df = data.frame(data=ceiling(runif(200,0,3)))

ggplot(df, aes(x=data)) + geom_bar(alpha=.2) + geom_histogram(alpha=.2)
```

```r
# CAT vs CAT
library(ggplot2)

df = data.frame(A=ceiling(runif(400,0,3)),B=ceiling(runif(400,3,6)))

ggplot(df, aes(x=A,y=B)) + geom_point(alpha=.2) + geom_jitter(alpha=.2)
```

```r
# Count of QUANT
library(ggplot2)

sex = rep(c('F','M'), c(100,500))  # Replace c(100,500) with each=500
height = c(rnorm(100,140,20), rnorm(500,180,20))
df = data.frame(sex, height)
df$sex = factor(df$sex, levels=c('M','F'), ordered=T)

# hist() and density() in base R
# Note: Histogram bins are STACKED, density plots illustrate proportion of individual CAT
ggplot(df, aes(x=height)) +
geom_histogram(aes(y = ..density.., fill=sex), alpha=I(.2)) +
geom_density(aes(fill=sex), alpha=I(.2))
```

```r
ggplot(df, aes(height)) + stat_ecdf(geom='point') +
geom_hline(yintercept=c(.25,.5,.75), lty=2) +
geom_vline(xintercept=fivenum(df$height)[2:4], lty=2)
```

```r
# CAT vs QUANT
ggplot(df, aes(sex, height)) +
geom_boxplot(aes(fill=sex), alpha=I(.2)) +
geom_jitter(aes(color=sex), alpha=I(.2))
```

```r
# QUANT vs QUANT
library(ggplot2)

x = -5:5
y = (x-1)*(x+2)
df = data.frame(x,y)

# qplot(x,y, data=df, geom=c('point','line')) +
# geom_hline(yintercept = 0, lty=2) +
# geom_vline(xintercept = 0, lty=2)

ggplot(df, aes(x,y)) +
geom_point() + geom_line() +
geom_hline(yintercept = 0, lty=2) +
geom_vline(xintercept = 0, lty=2)
```
