## Image format

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage import transform
```

```python
single_image = cv2.imread("data/000000000.jpg")
```

```python
plt.imshow(single_image)
plt.show()
```

```python
type(single_image)
```

```python
single_image.shape
```

```python
single_image = cv2.resize(single_image,(150, 150))
```

```python
single_image.shape
```

```python
gray_image = cv2.cvtColor(single_image, cv2.COLOR_RGB2GRAY)
```

```python
gray_image.shape
```

```python
flipped_image = np.fliplr(gray_image)
```

```python
plt.imshow(flipped_image)
plt.show()
```

```python
rotated_image = transform.rotate(single_image, angle=45)
```

```python
plt.imshow(rotated_image)
plt.show()
```

## Text format

```python
import pandas as pd
import nltk
import spacy
```

```python
reviews = pd.read_json("data/reviews_Digital_Music_5.json", lines=True)
```

```python
example_review = reviews["reviewText"].sample()
```

```python
example_review = str(example_review)
example_review
```

```python
type(example_review)
```

```python
sentences = nltk.sent_tokenize(example_review)
```

```python
sentences
```

```python
words = nltk.word_tokenize(sentences[0])
```

```python
english_stop_words = set(nltk.corpus.stopwords.words("english"))
cleaned_words = [word for word in words if word not in english_stop_words]
cleaned_words
```

Here you'll need to download the model: `python -m spacy download en_core_web_sm`

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(example_review)
```

```python
type(doc)
```

```python
for token in doc:
    print(token.text, token.pos_)
```
