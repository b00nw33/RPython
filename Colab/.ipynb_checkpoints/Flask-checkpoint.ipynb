```python
!pip install pyngrok
from flask import Flask, redirect, url_for
from pyngrok import ngrok

port_no = 5000
```

```python
app = Flask(__name__)
ngrok.set_auth_token("2L1fyu4MU1XBa8TM32RJ8kafjZI_6XbiVt8PKVezemnpDc6xz")
public_url =  ngrok.connect(port_no).public_url

# App begins below
@app.route("/")
def home():
  return f"<h1>Hello!</h1></br>I am running Flask on Google Colab!"

@app.route("/admin")
def admin():
  return redirect(url_for("user", name="Admin!!"))

@app.route("/<name>")
def user(name):
  return f"Hello {name}!"





# App continues below
print(f"To access the Global link please click {public_url}")

app.run(port=port_no)
```
