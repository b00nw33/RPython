```python
def collatz(number):
  lst=[]
  lst.append(number)
  while(number!=1):
    if(number%2==0):
      number=number//2
    else:
      number=number*3+1
    lst.append(number)
  return lst

import copy
li = [collatz(i) for i in range(1,5)]
li2 = copy.deepcopy(li)
li2[0] += ['test']
li2[1] *= 3
li2[1].remove(1)
li2[1].remove(1)
li
```

```python
import re

pattern = "([a-zA-Z0-9]+)@([a-zA-Z]+)\.(com|edu|net|org)"
new = r"\2 \1 \3"

while True:
  user = input()
  if user=="":
    break
  if (re.search(pattern, user)):
    print("Valid email entered")
    print(re.sub(pattern, new, user))
  else:
    print("Invalid input")
```

```python
spam = {'color':'red', 'age':42}

s = ','.join([f'The {k} is {v}' for k,v in spam.items()])

count={}
for i in s.lower():
  count.setdefault(i, 0)
  count[i] += 1

import pprint
pprint.pprint(count)
```
