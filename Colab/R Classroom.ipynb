```r
library(tibble)

df = tibble(x=1:100,
            y=21*x +3 + rnorm(100,0,50),
            xy = x*y, xsq = x^2
            )

s = colSums(df)
(100*s[3]-s[1]*s[2])/(100*s[4]-s[1]^2)
```

```r
fit = lm(y~x, df)
summary(fit)
```

```r
plot(df$x,df$y)
abline(fit)
```

```r
# Predict y given x
predict(fit, data.frame(x=0), interval = 'confidence')
```

```r
# FFT
library(tibble)
library(stats)

POINTS = 2^8

# delta_f = 10/POINTS
df = tibble(k=0:(POINTS-1), t=k/10, g_t=cos(2*pi*t)+cos(7*pi*t), mG = Mod(fft(g_t)), Hz=10*k/POINTS)[1:(POINTS/2),]

# plot(df$t,df$g_t,'o')
plot(df[['Hz']],df$mG,'b')
```

```r
plot(df$t,df$g_t,'l')
```
